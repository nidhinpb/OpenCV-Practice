```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse=cv2.imread("D:\Python CV_Course\Computer-Vision-with-Python\DATA\horse.jpg")
show_horse=cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB)

rainbow=cv2.imread("rainbow.jpg")
show_rainbow=cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB)

blue_bricks=cv2.imread("bricks.jpg")
show_bricks=cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dark_horse)
```

```python
plt.imshow(show_rainbow)
```

```python
plt.imshow(show_bricks)
```

```python
histbrick=cv2.calcHist([blue_bricks],channels=[0],mask=None,histSize=[256],ranges=[0,256])#BGR for opencv
plt.plot(histbrick)
```

```python
histhorse=cv2.calcHist([dark_horse],[0],None,[256],[0,256])
plt.plot(histhorse)
```

```python
img=blue_bricks
```

```python
color=("b","g","r")
for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)#color argument prints the histogram in the said color
    plt.xlim([0,256])
plt.title("HISTOGRAM FOR BLUE BRICKS")
```

```python
img2=dark_horse
color=("b","g","r")
for i,col in enumerate(color):
    histr=cv2.calcHist([img2],[i],None,[256],[0,256])
    plt.plot(histr,color=col)#color argument prints the histogram in the said color
    plt.xlim([0,256])
plt.title("HISTOGRAM FOR DARK HORSE")
```

```python
img3=show_horse
color=("b","g","r")
for i,col in enumerate(color):
    histr=cv2.calcHist([img3],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim=([0,100])
    plt.ylim=([0,50000])
plt.title("HISTOGRAM FOR DARK HORSE RGB FORM")
```
