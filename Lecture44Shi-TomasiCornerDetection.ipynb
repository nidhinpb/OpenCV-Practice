```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\flat_chessboard.png")
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
flat_chess_gray=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
real_chess=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\real_chessboard.jpg")
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
real_chess_gray=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(flat_chess_gray)
```

```python
plt.imshow(real_chess_gray)
```

```python
corners=cv2.goodFeaturesToTrack(flat_chess_gray,10,0.01,10)
corners=np.int0(corners)
for i in corners:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
plt.imshow(flat_chess)
```

```python
rcorners=cv2.goodFeaturesToTrack(real_chess_gray,100,0.01,10)#100 is the number of corners user wants to detect
rcorners=np.int0(rcorners)
for i in rcorners:
    x,y=i.ravel()
    cv2.circle(real_chess,(x,y),5,(255,0,0),-1)
plt.imshow(real_chess)
```
