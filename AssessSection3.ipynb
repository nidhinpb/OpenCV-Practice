```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img= cv2.imread('D:\Python CV_Course\Computer-Vision-with-Python\DATA\dog_backpack.jpg')
plt.imshow(img)
```

```python
fix_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
flip=cv2.flip(fix_img,0)
plt.imshow(flip)
```

```python
plt.imshow(fix_img)
```

```python
img_red=fix_img.copy()
cv2.rectangle(img_red,(200,380),(610,710),color=(255,0,0),thickness=8)
plt.imshow(img_red)
```

```python
plt.imshow(fix_img)
tri_img=fix_img.copy()
```

```python
vertices=np.array([[250,700],[400,400],[650,700]],dtype=np.int32)
vertices.shape
```

```python
pts=vertices.reshape(-1,1,2)
pts.shape
```

```python
cv2.polylines(tri_img,[pts],isClosed=True,color=(0,0,255),thickness=30)
cv2.fillPoly(tri_img,[pts],color=(0,0,255))
plt.imshow(tri_img)
```

```python
import numpy as np
import cv2
def cir_pic(events,x,y,flags,param):
    if events==cv2.EVENT_RBUTTONDOWN:
        cv2.circle(fix_img,(x,y),50,color=(0,0,255),thickness=10)    
cv2.namedWindow(winname="Puppy")
cv2.setMouseCallback("Puppy",cir_pic)
fix_img= cv2.imread('D:\Python CV_Course\Computer-Vision-with-Python\DATA\dog_backpack.jpg')
while True:
    cv2.imshow("Puppy",fix_img)
    if cv2.waitKey(20) & 0xFF==27:
        break
cv2.destroyAllWindows()
```
