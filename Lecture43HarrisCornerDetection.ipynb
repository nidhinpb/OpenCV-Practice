```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\flat_chessboard.png")
flat_chess_gray=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(flat_chess_gray)
```

```python
real_chess=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\real_chessboard.jpg")
real_chess_gray=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(real_chess_gray,cmap="gray")
```

```python
gray=np.float32(flat_chess_gray)
```

```python
dst=cv2.cornerHarris(gray,2,3,0.04)
dst=cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()]=[255,0,0]
plt.imshow(flat_chess)
```
