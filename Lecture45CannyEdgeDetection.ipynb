```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\sammy_face.jpg")
plt.imshow(img)
```

```python
edges=cv2.Canny(img,127,127)
plt.imshow(edges)
```

```python
edges=cv2.Canny(img,220,255)
plt.imshow(edges)
```

```python
med=np.median(img)
print(med)
```

```python
lower=int(max(0,0.7*med))
upper=int(min(255,1.3*med))
```

```python
edges=cv2.Canny(img,lower,upper)
plt.imshow(edges)
```

```python
edges=cv2.Canny(img,lower,upper+300)
plt.imshow(edges)
```

```python
edges=cv2.Canny(img,lower,upper+400)
plt.imshow(edges)
```

```python
blurred=cv2.blur(img,(3,3))
plt.imshow(blurred)
```

```python
blurred=cv2.blur(img,(3,3))
edges=cv2.Canny(blurred,lower,upper)
plt.imshow(edges)
```
