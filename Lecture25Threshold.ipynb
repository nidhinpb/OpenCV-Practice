```python
img=cv2.imread("D:\Python CV_Course\Computer-Vision-with-Python\DATA\crossword.jpg",0)
plt.imshow(img,cmap="gray")
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
show_pic(img)
```

```python
ret,th1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
ret,th1=cv2.threshold(img,180,255,cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

```python
#increasing threshold value brings intensity to black as above
#here we cannot find the best pic without experimenting the threshold values
#another alternative method to automatically achieves the threshold value based on the pixel
th2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended=cv2.addWeighted(th1,0.5,th2,0.5,10)
show_pic(blended)
```
