```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\car_plate.jpg")
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
pla_cas=cv2.CascadeClassifier(r"C:\Users\Kuttan\Downloads\Python CVCourse\Python CV_Course\Computer-Vision-with-Python\DATA\haarcascades\haarcascade_russian_plate_number.xml")
img_copy=new_img.copy()
rect_dime=pla_cas.detectMultiScale(img_copy,1.3,3)
for x,y,w,h in rect_dime:
    cv2.rectangle(img_copy,(x,y),(x+w,y+h),(255,0,0),10)
plt.imshow(img_copy)
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111)
ax.imshow(img_copy)
```

```python
x=rect_dime[0][0]
y=rect_dime[0][1]
w=rect_dime[0][2]
h=rect_dime[0][3]
print(x,y,w,h)
```

```python
slice=img_copy[y:y+h,x:x+w] 
plt.imshow(slice)
```
